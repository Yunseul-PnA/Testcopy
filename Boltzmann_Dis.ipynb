```python
h= const.h.value
c = const.c.value
kb = const.k_B.value
N= 1e11
lam = np.arange(100,15000,0.001) 
def Boltz(T):
    tt = (2*h*c**2/((lam*1e-10)**5))
    ex = ((h*c) / ( (lam*1e-10)*kb*T))
    aa = tt / (np.exp(ex) - 1)
    return aa

yy = Boltz(100000)
plt.plot(lam,yy)
# plt.xlim(0,2000)
# plt.yscale('log')
print(len(yy))
print(lam[np.argmax(yy)])
```
