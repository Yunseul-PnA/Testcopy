```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd

# 데이터 로드
tt = pd.read_csv(r'/home/jin/cloudy_Test/Young PNe.csv')
Tem, Lum = tt['Teff(K)'].to_numpy(), tt['L (Sun)'].to_numpy()
Ha_Hb = np.array([7.39, 2.8, 5.02, 42.21, 5.25, 17.28, 6.38, 3.58, 3.6, 7.83, 5.32, 4.03])
name = tt['Name'].to_numpy()

GroupR = np.array(['NGC 6741', 'NGC 6881', 'NGC 6886', 'NGC 6884'])

# Backend 설정
import matplotlib
# matplotlib.use('Qt5Agg')  # 또는 'Qt5Agg'

# 3D 플롯 생성
fig = plt.figure()
ax = plt.axes(projection='3d')

x = np.log10(Tem)
y = np.log10(Lum)

for ii, nn in enumerate(name):
    if nn == 'Hen2-447':
        print('hi')
    else:
        if nn in GroupR:
            ax.plot3D(y[ii], x[ii], Ha_Hb[ii], 'ro')  # Red points
            ax.text(y[ii], x[ii], Ha_Hb[ii], nn)
        else:
            ax.plot3D(y[ii], x[ii], Ha_Hb[ii], 'bo')  # Blue points
            ax.text(y[ii], x[ii], Ha_Hb[ii], nn)

ax.set_xlabel('L')
ax.set_ylabel('T')
ax.set_zlabel('Ha_Hb')

# Interactive 모드 활성화
# plt.ion()

# 플롯 출력
plt.show()
```
